```python
from __future__ import print_function

import sys

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import scipy.stats.stats as scss
```

生成模型数据，其中x1，x2为不相关的变量；x1，x3强相关。

```python
def generateData(n):
    """
    生成模型数据，其中x1，x2为不相关的变量；x1，x3强相关

    参数：
    ----
    n : int，数据重复次数
    """
    data = []
    np.random.seed(2046)
    for i in range(0, 3):
        for j in range(0, 3):
            data.append({"x1": i, "x2": j})
    data = pd.DataFrame(data * n)
    # 生成强相关自变量
    data["x3"] = data["x1"] + np.random.random(9 * n)
    # 生成被预测值
    error = 0.1 * np.random.random(9 * n)
    data["y"] = 0.7 * data["x1"] - 1.1 * data["x2"] + 0.3 * data["x3"] + error
    return data
data = generateData(3)
```

```python
data.head()
```

** 可视化数据并计算各变量之间的相关系数矩阵**

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False # 正常显示负号

axes = scatter_matrix(data, alpha=1, figsize=(9,9), diagonal='kde', range_padding=0.9)

# 计算各变量之间的皮尔逊相关系数
corr = data.corr(method='pearson').values
for i, j in zip(*plt.np.triu_indices_from(axes, k=1)):
    axes[i,j].annotate("%s: %.3f" % ("相关系数".decode('utf-8'), corr[i,j]),(0.5, 0.9),xycoords='axes fraction',ha='center',va='center')
```

## 多重共线性效应
** 首先使用不相关的$x_1$和$x_2$搭建3个不同的线性模型**

```python
def model(data, formula):
    
    model = sm.OLS.from_formula(formula,data=data)
    re = model.fit()
    print(re.summary())
    return re
```

```python
formula = 'y ~ x1'
model(data, formula)
```

```python
formula = 'y ~ x2'
model(data, formula)
```

```python
formula = 'y ~ x1 + x2'
model(data, formula)
```

** 然后使用强相关的$x_1$和$x_3$同样搭建3个模型模型**

```python
formula = 'y ~ x3'
model(data, formula)
```

```python
formula = 'y ~ x1 +x3'
re = model(data, formula)
```

根据上述结果可以看出，$x_1$和$x_3$被模型误判为不显著，于是我们可以试着利用F-test检验$x_1$和$x_3$同时不显著（即联合假设检验）

```python
print(re.f_test(['x1=0','x3=0']))
```

拒绝$x_1$和$x_3$同时不显著的结论，那么我们就得到了当$x_1$和$x_3$同时出现在模型中时，被模型判断为不显著，但是采用联合假设检验$x_1$和$x_3$同时不显著时，得到的结论却是否定的。这也是检验多重共线性效应的一种方法。

## 共线性的解决方法

### 结构化共线性

** 将变量的中心重置为0**

```python
plt.scatter(data['x3'],data['x3']**2)
```

```python
data['x3_square']=data['x3']**2
data['center_x3']=data['x3']-data['x3'].mean()
data['center_x3_square'] = data['center_x3']**2
# 重置前
formula = 'y ~ x3 +x3_square'
re = model(data, formula)
```

```python
plt.scatter(data['center_x3'],data['center_x3_square'])
```

```python
# 重置后
formula = 'y ~ center_x3 + center_x3_square'
re = model(data, formula)
```

** one-way ANOVA分析的 $\eta^2$**

```python
def generateData():
    """
    """
    np.random.seed(2046)
    d1 = np.random.normal(5, 5, 10)
    d2 = np.random.normal(5, 5, 10)
    groups = ["d1"] * 10 + ["d2"] * 10
    d = {"A": np.concatenate((d1, d2)), "B": groups}
    data = pd.DataFrame(d)
    return data
data = generateData()
data.head()
```

```python
data.boxplot(column="A",by="B", grid=False)
```

```python
def oneWayANOVA(data):
    """
    计算定量变量A与定性变量B之间的eta squared

    参数：
    ----
    data : DataFrame, 包含变量A和变量B
    """
    re = sm.OLS.from_formula('A~B',data=data).fit()
    aovTable = sm.stats.anova_lm(re, typ=2)
    print(aovTable)
    etaSquared = aovTable['sum_sq'][0]/(aovTable['sum_sq'][0]+aovTable['sum_sq'][1])
    return etaSquared
```

```python
oneWayANOVA(data)
```

** 这个结果还未理解？？**

```python
def increaseDataSet():
    """
    通过增加数据量来解决共线性问题
    """
    index = []
    x1Std = []
    x1 = []
    vif = []
    for i in range(1, 150):
        data = generateData(i)
        formula = 'y ~ x1 + x3'
        re = model(data, formula)
        k = re.cov_params()
        index.append(i)
        x1Std.append(np.sqrt(k.loc["x1", "x1"]))
        x1.append(re.params["x1"])
    # 为在Matplotlib中显示中文，设置特殊字体
    plt.rcParams['font.sans-serif']=['SimHei']
    # 正确显示负号
    plt.rcParams['axes.unicode_minus']=False
    fig = plt.figure(figsize=(6, 6), dpi=80)
    ax = fig.add_subplot(1, 1, 1)
    # 在Python3中，str不需要decode
    if sys.version_info[0] == 3:
        ax.plot(index, x1, "r-.", label=u'%s' % "a1的估计值")
        ax.plot(index, x1Std, "b", label=u'%s' % "a1标准差估计值")
    else:
        ax.plot(index, x1, "r-.", label=u'%s' % "a1的估计值".decode("utf-8"))
        ax.plot(index, x1Std, "b", label=u'%s' % "a1标准差估计值".decode("utf-8"))
    legend = plt.legend(loc=4, shadow=True)
    plt.show()
increaseDataSet()
```
